```python
#Imports
%matplotlib inline
import pandas
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import numpy as np

pandas.set_option('display.max_columns', None)  
pandas.set_option('display.expand_frame_repr', False)
pandas.set_option('display.precision', 3)
```

```python
data = pandas.read_csv('./data/Train.csv', sep=',', na_values=".")
```

```python
data.describe(include='all')
```

```python
#Comprovar quins camps tenen valors nulls
data.isnull().any()
```

```python
values = {"Alcohol Consumption": 'Never', "Chronic Diseases": 'No', "Medication Use": 'No', "Family History": 'No Disease', "Education Level": 'No Education'}
data = data.fillna(value=values)
data.isnull().any()
```

```python
avis = data[data['Age_group'] == 2]
avis.boxplot(meanline=False,showmeans=True,rot=90,return_type='axes')
```

```python
adults = data[data['Age_group'] == 1]
adults.boxplot(meanline=False,showmeans=True,rot=90,return_type='axes')
```

```python
joves = data[data['Age_group'] == 2]
joves.boxplot(meanline=False,showmeans=True,rot=90,return_type='axes')
```
